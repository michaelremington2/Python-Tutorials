```python
grades= [100*rd.random() for _ in range(100)]
decile = lambda grade:grade // 10*10
histogram = Counter(decile(grade) for grade in grades)
x=histogram.keys()
y=histogram.values()
error=s.stdev(grades)/math.sqrt(len(grades))

m.bar(x,y,9,yerr=error)
m.axis([-5,105,0,20])
m.xticks([10*i for i in range(11)])
m.show()
```

```python
variance=[1,2,4,8,16,128,256,25]
bias_squared=sorted(variance, reverse=True)
total_error =[x+y for x,y in zip(variance,bias_squared)]
xs = [i for i, _ in enumerate(variance)]
## multiple calls to plot overlays multiple charts on same plot
plt.plot(xs,variance,'g-', label='variance')
plt.plot(xs,bias_squared,'c-', label='bias_squared')
plt.plot(xs,total_error,'b:', label='total_error')

plt.legend(loc=9)
plt.axis([0,8,7,275])
plt.show()
```
